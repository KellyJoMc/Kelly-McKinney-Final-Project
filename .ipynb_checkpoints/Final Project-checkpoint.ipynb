```python
import pandas as pd
import numpy as np
```

# Data wrangling

## changing created_at and last_move_at columns to datetime data type

```python
chess = pd.read_csv('chess.csv')

df_times = chess[['created_at','last_move_at']].copy()

df_times['created_at_dt'] = pd.to_datetime(df_times['created_at']/1000, 
                                           unit='s', 
                                           origin='unix')

df_times['last_move_at_dt'] = pd.to_datetime(df_times['last_move_at']/1000, 
                                             unit='s', 
                                             origin='unix')

df_times.tail(6)
```

## Merging new dataset with original dataset

```python
chess1 = pd.concat([chess, df_times], axis=1)
chess1.head()
```

## Dropping original created_at and last_move_at columns

```python
chess1.drop(['created_at', 'last_move_at'], axis=1)
```
